*(2) Write a program in R to find factorial of a number by taking input from user. Please
print error message if the input number is negative.*

```python
# factorial of a number 
# this will read input and convert it into string 
var = readline(prompt = "Enter a number") 
n = as.integer(var)

if(n<0){
  cat("invalid input")
}else{
  i = 1
  facto = 1 
  
  while(i<=n)
  {
    facto = facto * i
    i = i+1
  }
  
  cat("factorial is: ", facto)
}
```

(2) A room has n people, and each has an equal chance of being born on any of the 365
days of the year. (For simplicity, we’ll ignore leap years). What is the probability
that two people in the room have the same birthday? <br>
(a) Use an R simulation to estimate this for various n. <br>
(b) Find the smallest value of n for which the probability of a match is greater than
.5.

```python
%%R
# (a)
# n is the number of people
n = 100
prod = 1

# finding the probability that none of the people have the same birthday
for(i in 0:(n-1))
{
  prod = prod * (1-(i/365))
}
prob = 1 - prod
cat("probability = ", prob)
cat("\n")

# (b)
n = 2
while (n >= 2){
    prod = 1
    for (i in 1:(n-1))
    {
        prod = prod * (1-(i/365))
    }
    prob = 1-prod
    if (prob > 0.5){
        break
    }
    n = n + 1
}
# cat("smallest value of n : ", n)
```

(3) Write an R function for computing conditional probability. Call this function to do
the following problem:
suppose the probability of the weather being cloudy is 40%. Also suppose the probability of rain on a given day is 20% and that the probability of clouds on a rainy day
is 85%. If it’s cloudy outside on a given day, what is the probability that it will rain
that day?


```python
%%R
# solve this using conditional probability in Bayes theorem, 

BayesTheorem = function(pRain, pCloudy, pCloudyRain){
  pRainCloudy = (pRain * pCloudyRain) / pCloudy
  return(pRainCloudy)
}


pRain = 0.2
pCloudy = 0.4
pCloudyRain = 0.85

prob = BayesTheorem(pRain, pCloudy, pCloudyRain)
cat("Probability : ", prob)
```

```python
%%R
getMode = function(data){
  uniData = unique(data) 
  # creates a vector with only the unique elements from the given dataset
  
  # uniData[which.max(tabulate(match(data,uniData)))]
  uniData[which.max(tabulate(   match(data,uniData)   ))]
  
  # 1. match(data,unidata)
  # 2. tabulate()
  # 3. which.max()
  # 4. uniData[]
  
  # match() function matches the data from the unique vector to the given vector and returns the index at which it finds that data
  # The match() function in R is used to: Return the index position of the first element present in a vector with a specific value
  # tabulate() function will create a frequency table of the given indices
  # which.max() will return the maximum value of the frequency table created by tabulate
}

data = c(2,1,2,3,1,2,3,4,1,5,5,3,2,3)
result = getMode(data)
print(result)




```

 Write a r-code <br>
(i) To find the 95th percentile of the F-distribution with (10, 20) degrees of freedom.<br>
(ii) To calculate the area under the curve for the interval [0, 1.5] and the interval [1.5, +∞) of
a F-curve with v1 = 10 and v2 = 20 (USE pf()). <br>
(iii) To calculate the quantile for a given area (= probability) under the curve for a F-curve 
with v1 = 10 and v2 = 20 that corresponds to q = 0.25, 0.5, 0.75 and 0.999. (use the qf()) <br>
(iv) To generate 1000 random values from the F-distribution with v1 = 10 and v2 = 20 (use
rf())and plot a histogram.

```python
# We use the qf() to calculate the quantile for a given area (= probability) under the curve for a F-curve
A <- qf(.95,df1=10,df2=20)
print(A)

# We use the pf() to calculate the area under the curve for the interval
B1 <- pf(1.5,df1=10,df2=20,lower.tail = TRUE)
B2 <- pf(1.5,df1=10,df2=20,lower.tail = FALSE)
print(B1)
print(B2)

# We use the qf() to calculate the quantile for a given area (= probability) under the curve for a F-curve
C1 = qf(0.25,df1=10,df2=20)
C2 = qf(0.5,df1=10,df2=20)
C3 = qf(0.75,df1=10,df2=20)
C4 = qf(0.999,df1=10,df2=20)
print(C1)
print(C2)
print(C3)
print(C4)

# We use rf()function to generate 100,000 random values from the F-distribution
D <- rf(n = 1000, df1=10,df2=20)
hist(D)
print(D)
```
